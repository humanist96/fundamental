```python
from ckonlpy.tag import Postprocessor
from ckonlpy.tag import Twitter
from ckonlpy.utils import load_wordset
from ckonlpy.utils import load_replace_wordpair
from ckonlpy.utils import load_ngram

#passwords = {'아이오아이', ('정말', 'Noun')}
passwords = {}
stop_words = ['전국','주택', '건설', '사업','부동산', '가구', '사흘', '내달', '이번','단지', '사전','지난해','까지', '면서', '올해', '지난', '지역', '위해', '현재', '최근', '부터', '고의', '경우', '달리', '적지',
              '모두', '일부','모든','보통','또한','관련','주로','구체','때문','로부터', '당시','지금', '도록', '자동', '가까이', '기타',\
             '다음','주로','대부분','취할','구성은','대해','통해','마련','따라서','제대로','다시', '측면', '먼저', '적극', '대한', \
             '조차', '마저', '이후', '벌써', '매우', '결코', '간의', '방안', '분석', '이나', '가장', '라며', '다만', '만원', '었으', '롭다', '대비',\
              '전고', '나타']
replace = {'감정': '감정가', '낙찰':'낙찰가', '서울시':'서울'}
#replace = {'감정': '감정가', ('이뻐', 'Adjective'): '예쁘다'}
ngrams =[(('감정가', '가'), 'Noun'), (('낙찰', '가'), 'Noun')]
# 새로운 단어 추가
new_nouns = ['전세사기', '주택 건설 사업', '현대건설', '포스코이엔씨', '대우건설', '호반건설', '금강주택']
# with open('preprocess/dictionary.txt', encoding='utf8') as fd:
#     for line in fd:
#         new_nouns.append(line.strip('\n'))

twitter = Twitter()
twitter.add_dictionary(new_nouns, 'Noun')
passtags = {'Noun'}

postprocessor = Postprocessor(
    base_tagger = twitter,
    stopwords = stop_words,
    passtags = passtags,
    replace = replace,
    ngrams = ngrams
)
```

```python
# 커스터마이징 tiweeter 사용
def tokenize_list_custom(text): #text를 통으로 주느냐, 하나씩 주느냐
  t_nouns = postprocessor.pos(text)
  words = []
  for word, pos in tqdm(t_nouns):
    if pos == 'Noun' and (len(word) > 1) :  # 이 경우에는 형용사, 동사, 명사만 남김
        words.append(word)
  #words_str = ' '.join(words)
  return words

#전체 텍스트에 대한 wordlist
words = tokenize_list_custom(only_texts)
```
